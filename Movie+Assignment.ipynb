-  ### Subtask 3.6: Find popular genres

You might have noticed the `genres` column in the dataframe with all the genres of the movies seperated by a pipe (`|`). Out of all the movie genres, the first two are most significant for any film.

1. Extract the first two genres from the `genres` column and store them in two new columns: `genre_1` and `genre_2`. Some of the movies might have only one genre. In such cases, extract the single genre into both the columns, i.e. for such movies the `genre_2` will be the same as `genre_1`.
2. Group the dataframe using `genre_1` as the primary column and `genre_2` as the secondary column.
3. Find out the 5 most popular combo of genres by finding the mean of the gross values using the `gross` column and store them in a new dataframe named `PopGenre`.

```python
# Write your code for extracting the first two genres of each movie here
#first, split the genre column with a pipe('|') as separator
split_genre = movies['genres'].str.split('|',  expand=True)
# did we get a column which has a null value? or only one genre type?, fill the genre2 with genre1
split_genre[1].fillna(split_genre[0], inplace=True)
# drop the rest of columns other than 0 and 1, this is required to clean up the other genre categories
split_genre.drop(range(2,len(split_genre.columns)), axis=1, inplace=True)
#rename the respective columns
split_genre.rename(columns={0:'genre_1',1:'genre_2'})
# inspect the genres now
split_genre.info()
```

```python
#assign back the columns to original DF
# add column, genre_1
movies['genre_1'] = split_genre[0]
# add column, genre_2
movies['genre_2'] = split_genre[1]
#drop column 'genre' as it is of no use
movies.drop('genres', inplace=True, axis=1)
movies.info()
```

```python
 # Write your code for grouping the dataframe here
movies_by_segment = movies.groupby(['genre_1', 'genre_2'])
# fetch the movies with highest gross in the respective genre_1 and genre_2
movies_by_segment = movies_by_segment['gross'].mean().sort_values(ascending=False)
#inspect the movies_by_segment
movies_by_segment
```

```python
 # Write your code for getting the 5 most popular combo of genres here
PopGenre = pd.DataFrame(movies_by_segment.head(5))
#reset index and inspect the PopGenre DataFrame.
PopGenre.reset_index()
```

```python
# Write the code for calculating decade here
#first convert the data type of the title_year column, to int
movies.title_year = movies.title_year.astype(np.int)
#do a sanity check to ensure we have the title_year as ints.
movies.title_year.head(5)
#now, add the decade columns
movies['decade'] = 10 * (movies.title_year // 10)
#sanity check, verify decade column presence
movies.decade.head(5)
```

```python
#sort the dataframe by column decade
movies.sort_values('decade', ascending = True, inplace = True)
#inspect for sanity
movies.head(5)
```

```python
#group the movies by decade
df_by_decade_group = movies.groupby('decade')
```

```python
# Write your code for creating the data frame df_by_decade here 
df_by_decade = pd.DataFrame(df_by_decade_group['num_voted_users'].sum())
#do a sanity check of the decade, first reset index!
df_by_decade.reset_index(inplace=True)
df_by_decade
```

```python
# Write your code for plotting number of voted users vs decade
# make up a figure area for plotting the graphs, the figsize indicate the row x col inches(default).
plt.figure(figsize=(12,6))
graph = sns.barplot(x='decade',y='num_voted_users',data=df_by_decade,estimator=np.median)
plt.show()
```

```python
#alternatively, we could scale the 'y-axis' to log (as the range is high) to make the graph uncluttered
plt.figure(figsize=(12,6))
graph = sns.barplot(x='decade',y='num_voted_users',data=df_by_decade,estimator=np.median)
graph.set_yscale('log')
plt.show()
```
